```python
import torch
```

```python
dtype = torch.FloatTensor
```

```python
bs, N_in, N_out, N_h = 64, 1000, 10, 100
```

```python
x = torch.randn(bs, N_in).type(dtype)
y = torch.randn(bs, N_out).type(dtype)
```

```python
w1 = torch.randn(N_in, N_h).type(dtype)
w2 = torch.randn(N_h, N_out).type(dtype)
```

```python
lr = 1e-6
```

```python
for t in range(1000):
    
    # Forward pass
    h = x.mm(w1)
    h_relu = h.clamp(min=0)
    y_pred = h_relu.mm(w2)
    
    # Compute the loss
    loss = (y_pred - y).pow(2).sum()
    print('loss', loss)
    
    # Backward pass
    grad_y_pred = 2*(y_pred-y)
    grad_w2 = h_relu.t().mm(grad_y_pred)
    grad_h_relu = grad_y_pred.mm(w2.t())
    grad_h = grad_h_relu.clone()
    grad_h[h<0] = 0
    grad_w1 = x.t().mm(grad_h)
    
    # Update the weights
    w1 -= grad_w1 * lr
    w2 -= grad_w2 * lr
```

### Using Variables

```python
import torch
from torch.autograd import Variable
```

```python
dtype = torch.FloatTensor
```

```python
bs, N_in, N_out, N_h = 64, 1000, 10, 100
```

```python
x = Variable(torch.randn(bs, N_in).type(dtype), requires_grad=False)
y = Variable(torch.randn(bs, N_out).type(dtype), requires_grad=False)
```

```python
w1 = Variable(torch.randn(N_in, N_h).type(dtype), requires_grad=True)
w2 = Variable(torch.randn(N_h, N_out).type(dtype), requires_grad=True)
```

```python
lr=1e-6
```

```python
for t in range(500):
    # forward pass
    #h = x.mm(w1)
    #h_relu = h.clamp(min=0)
    #y_pred = h_relu.mm(w2)
    y_pred = x.mm(w1).clamp(min=0).mm(w2)
    
    # Compute the loss
    loss = (y_pred-y).pow(2).sum()
    print(t, loss.data[0])
    
    # Backward pass
    loss.backward()
    
    # Upgrade the weights
    w1.data -= lr * w1.grad.data
    w2.data -= lr * w2.grad.data
    
    w1.grad.data.zero_()
    w2.grad.data.zero_()
```
