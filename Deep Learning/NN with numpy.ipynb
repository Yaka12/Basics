```python
import numpy as np
```

```python
BS, N_in, N_out, N_h = 64, 1000, 10, 100
```

```python
x = np.random.randn(BS, N_in)
y = np.random.randn(BS, N_out)
```

```python
w1 = np.random.randn(N_in, N_h)
w2 = np.random.randn(N_h, N_out)
```

```python
lr = 1e-6
```

```python
for t in range(1000):
    # forward pass
    h = x.dot(w1)
    h_relu = np.maximum(h,0)
    y_pred = h_relu.dot(w2)
    
    # loss
    loss = np.square(y_pred - y).sum()
    print('loss:', loss)
    
    # Backprop
    grad_y_pred = 2. *(y_pred - y)
    grad_w2 = h_relu.T.dot(grad_y_pred)
    grad_h_relu = grad_y_pred.dot(w2.T)
    grad_h = grad_h_relu.copy()
    grad_h[h < 0] = 0
    grad_w1 = x.T.dot(grad_h)
    
    
    # Update the weights
    w1 -= lr*grad_w1
```
