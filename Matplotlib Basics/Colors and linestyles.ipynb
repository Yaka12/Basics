## Import the libraries

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Create something to plot

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

## Colors

```python
plt.plot(x, x**2, 'b.-') # blue line with dots
plt.plot(x, x**3, 'g--') # green dashed line
plt.show()
```

```python
plt.plot(x, x**2, color="blue", alpha=0.5) # half-transparant
plt.plot(x, x**3, color="#DC1155")        # RGB hex code
plt.show()
```

## Linestyle

```python
plt.plot(x, x**2, color="red", linewidth=2.5)
plt.plot(x, x**3, color="red", linewidth=0.75)
plt.show()
```

```python
plt.plot(x, x**2, color="green", lw=2, linestyle='-')
plt.plot(x, x**3, color="green", lw=2, ls='-.')
plt.show()
```

```python
plt.plot(x, x**2, color="blue", lw=2, ls='-', marker='s')
plt.plot(x, x**3, color="blue", lw=2, ls='--', marker='o')
plt.show()
```

```python
plt.plot(x, x**2, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
plt.plot(x, x**3, color="purple", lw=1, ls='-', marker='s', markersize=8, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
plt.show()
```
