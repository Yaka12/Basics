## Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create two curves

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

## Create the first subplot

The subplot will have 2 rows and 1 column

```python
#subplot(nrows, ncols, plot_number)
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Sine')
```

## Create the second subplot

```python
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')
```

## Show the figure

```python
plt.show()
```
