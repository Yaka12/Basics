## Import the libraries

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Create something to plot

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

## Add labels to the curves and plot the legend

```python
fig, ax = plt.subplots()
plt.plot(x, x**2, label="x**2")
plt.plot(x, x**3, label="x**3")
plt.legend()
plt.show()
```

### Choose the position of the legend

```python
plt.plot(x, x**2, label="x**2")
plt.plot(x, x**3, label="x**3")
plt.legend(loc=1) # upper right corner
plt.legend(loc=2) # upper left corner
plt.legend(loc=3) # lower left corner
plt.legend(loc=4) # lower right corner

# Setting loc=0 to use the optimal position
plt.legend(loc=0) 
plt.show()
```

## Add a title

```python
ax.set_title("title")
fig
```

## Add axes labels

```python
ax.set_xlabel("x")
ax.set_ylabel("y");
fig
```

### Twin axes

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```
