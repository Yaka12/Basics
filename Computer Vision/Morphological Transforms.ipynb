```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'sample.png'
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.axis('off')
```

## Create the kernel

```python
kernel = np.ones((3,3),np.uint8)
```

## Erosion

```python
erosion = cv2.erode(img,kernel,iterations = 1)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(erosion, cmap='gray')
plt.title('Erosion')
plt.axis('off')
plt.show()
```

## Dilation

```python
dilation = cv2.dilate(img,kernel,iterations = 1)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(dilation, cmap='gray')
plt.title('Dilation')
plt.axis('off')
plt.show()
```

## Opening

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(opening, cmap='gray')
plt.title('Opening')
plt.axis('off')
plt.show()
```

## Closing

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(closing, cmap='gray')
plt.title('Closing')
plt.axis('off')
plt.show()
```

## Morphological Gradient

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(gradient, cmap='gray')
plt.title('Gradient')
plt.axis('off')
plt.show()
```

## Top Hat

```python
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(tophat, cmap='gray')
plt.title('Top Hat')
plt.axis('off')
plt.show()
```

## Black Hat

```python
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(blackhat, cmap='gray')
plt.title('Black Hat')
plt.axis('off')
plt.show()
```
