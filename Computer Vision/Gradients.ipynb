```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'sample.png'
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.axis('off')
```

## Sobel

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X')
plt.axis('off')
plt.show()
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')
plt.show()
```

## Laplacian

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian')
plt.axis('off')
plt.show()
```
