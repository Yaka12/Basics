```python
import cv2
import matplotlib.pyplot as plt
```

```python
path = 'lena.png'
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

## Averaging

```python
blur = cv2.blur(img, (9,9))
```

```python
plt.figure(figsize=(8,8))
plt.subplot(121)
plt.imshow(img)
plt.title('original')
plt.axis('off')
plt.subplot(122)
plt.imshow(blur)
plt.title('blurred')
plt.axis('off')
plt.show()
```
